```python
###############################################################################################
""" Creating User Database"""
import sqlite3

# Connect to SQLite database (creates one if doesn't exist)
conn = sqlite3.connect('Data/users_data.db')
cursor = conn.cursor()

# Create a table for user profiles if not exists
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Users(
        id INTEGER PRIMARY KEY,
        username TEXT,
        age INTEGER,
        gender TEXT,
        fav_entertainment TEXT,
        least_fav_entertainment TEXT,
        likes TEXT,
        dislikes TEXT,
        movie_watching_freq TEXT,
        show_watching_freq TEXT,
        reading_freq TEXT
    )
''')

# Function to insert user profile into the database
def insert_user_profile(username, age, gender, fav_entertainment, least_fav_entertainment, likes, dislikes, movie_watching_freq, show_watching_freq, reading_freq):
    try:
        conn = sqlite3.connect('Data/users_data.db')
        cursor = conn.cursor()

        cursor.execute('''
            INSERT INTO Users (username, age, gender, fav_entertainment, least_fav_entertainment, likes, dislikes, movie_watching_freq, show_watching_freq, reading_freq)
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        ''', (username, age, gender, fav_entertainment, least_fav_entertainment, likes, dislikes, movie_watching_freq, show_watching_freq, reading_freq))
        
        conn.commit()
        conn.close()  # Close the connection after committing changes
        print("User profile inserted successfully.")
    except sqlite3.Error as e:
        print(f"Error inserting user profile: {e}")
```

```python
def extract_descriptions(text):
    descriptions = []
    # Split the text by lines
    lines = text.split('\n')
    # Find lines starting with '- "'
    for line in lines:
        if line.startswith('- "'):
            # Extract the description between the double quotes
            description = line.split(': ')[1]
            descriptions.append(description)
    return descriptions

text = '''
Fantastic! In that case, here are some romantic comedy movie recommendations for you:

<<<MOVIES_ON>>>
- "Crazy, Stupid, Love": A charming and hilarious film about love, relationships, and self-discovery.
- "The Proposal": A delightful romantic comedy with a mix of humor and heartwarming moments.
- "10 Things I Hate About You": A modern retelling of Shakespeare's "The Taming of the Shrew" with a perfect blend of romance and comedy.
- "Silver Linings Playbook": A unique and heartfelt story about two individuals finding love and redemption amidst their personal struggles.
- "When Harry Met Sally": A timeless classic that explores the complexities of friendship and love.

I hope you find these recommendations enjoyable! Let me know if there's anything else I can assist you with.
'''

descriptions = extract_descriptions(text)
print(descriptions)
```
